Join Types

```python
happy_scores =happy_scores[happy_scores.year > 2019]
happy_scores
```

```python
pop = pd.read_html('https://en.wikipedia.org/wiki/List_of_countries_by_population_(United_Nations)')[0]
pop = pop[pop['UN statistical subregion[4]'] == 'Northern America']
pop
```

```python
happy_scores.merge(pop, how='inner', left_on='country_name', right_on='Country / Area')
```

```python
happy_scores.merge(pop, how='outer', left_on='country_name', right_on='Country / Area')
```

```python
happy_scores.merge(pop, how='left', left_on='country_name', right_on='Country / Area')
```

```python
happy_scores.merge(pop, how='right', left_on='country_name', right_on='Country / Area')
```
